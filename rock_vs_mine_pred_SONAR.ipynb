```python
import pandas as pd
df = pd.read_csv('/content/Copy of sonar data.csv',header=None)
df = df.sample(n = len(df) , random_state=2)
df.head()
```

```python
# Encoding the Final Column
df.replace({60 : {'M': 0, 'R' : 1  }} , inplace = True)
# In the Final Column '0' means Mine and '1' means Rock
```

```python
df.head()
```

```python
x = df.drop(60 , axis =1 )
y = df[60]
```

```python
print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=1 , stratify = y)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train , y_train)
train_model_pred = lr.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
train_accuracy_score = accuracy_score(y_train , train_model_pred)
print(train_accuracy_score)
```

```python
# Now checking Accuracy score for Test Data
test_model_pred = lr.predict(x_test)
test_accuracy_score= accuracy_score(y_test , test_model_pred )
print(test_accuracy_score)
```

```python
# Building a Predictive Model
import numpy as np
def model_pred(data):
  data = np.asarray(data)
  data = data.reshape(1,-1)
  prediction = lr.predict(data)
  if prediction[0] == 0:
    print('There is a MINE, Be Careful ')
  else:
    print('Chill Its Just a ROCK')
```
