```python
import pandas as pd
```

```python
df = pd.read_csv('/content/heart_disease_data.csv')
df_shuffled = df.sample(n = len(df) , random_state=3)
df.head()

```

```python
x = df.drop('target' , axis = 1)
y = df['target']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,  x_test, y_train, y_test = train_test_split(x , y , test_size = 0.2 ,random_state = 2, stratify=y) #stratify helps to distribute data in equal proportion of target(0 and 1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr
```

```python
lr.fit(x_train , y_train)
```

```python
train_model_pred = lr.predict(x_train)
```

```python
# for classification problem we'll be using Accuracy Score
from sklearn.metrics import accuracy_score
acs_train = accuracy_score(y_train , train_model_pred)
print(f"accuracy score for training data is {acs_train}")
```

```python
# Now calculating accuracy score on Test Data
train_model_pred = lr.predict(x_test)
acs_test = accuracy_score(y_test , train_model_pred)
print(f"accuracy score for test data is {acs_test}")
```
